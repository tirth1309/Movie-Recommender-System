```python
import statistics

def print_descriptive_statistics(data):
        
    try:
        mean = statistics.mean(data)
        print(f"Mean: {mean}")
    except statistics.StatisticsError:
        print("Mean: Data is empty.")
    
    try:
        median = statistics.median(data)
        print(f"Median: {median}")
    except statistics.StatisticsError:
        print("Median: Data is empty or not valid.")
    
    try:
        mode = statistics.mode(data)
        print(f"Mode: {mode}")
    except statistics.StatisticsError:
        print("Mode: No unique mode found or data is empty.")
    
    try:
        data_range = max(data) - min(data)
        print(f"Range: {data_range}")
    except ValueError:
        print("Range: Data is empty or not valid.")
    
    try:
        variance = statistics.variance(data)
        print(f"Variance: {variance}")
    except statistics.StatisticsError:
        print("Variance: Data is empty or variance requires at least two data points.")


```

```python
m2_users = []
import json
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Read the JSON file into a Python dictionary
with open('../data/drift/milestone2-3M.json', 'r') as file:
    data = json.load(file)

# Steps 2 and 3: Extract and filter movie ratings
all_ratings = []
for user_id, user_data in data.items():
    movies = user_data["movies_watched"]
    for movie, details in movies.items():
        rating = details["rating"]
        if rating is not None:  # Filtering out 'null' ratings
            all_ratings.append(rating)
            m2_users.append(user_id)

# Step 4: Plot the distribution of the ratings
plt.figure(figsize=(8, 6))
sns.histplot(all_ratings, bins=5, kde=True)
plt.title('Distribution of User Rating')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()

print_descriptive_statistics(all_ratings)
```

```python
def analyze_data_from_json(file_path, detail_key, limit, bins):
    """
    Reads a JSON file into a Python dictionary, filters data based on a given detail key,
    plots the distribution, and prints descriptive statistics.
    
    Parameters:
    - file_path: The file path to the JSON file.
    - detail_key: The key in the JSON data details to analyze (e.g., "minutes").
    """
    # Step 1: Read the JSON file
    with open(file_path, 'r') as file:
        data = json.load(file)
    
    # Steps 2 and 3: Extract and filter data
    all_details = []
    for user_id, user_data in data.items():
        movies = user_data.get("movies_watched", {})
        for movie, details in movies.items():
            detail_value = details.get(detail_key)
            if detail_value is not None and len(all_details) < limit:
                all_details.append(detail_value)            
    
    print(f"Total data points extracted: {len(all_details)}")
    # Step 4: Plot the distribution
    plt.figure(figsize=(8, 6))
    sns.histplot(all_details, bins=bins, kde=True)
    plt.title(f'Distribution of {detail_key} Across Movies')
    plt.xlabel(detail_key.capitalize())
    plt.ylabel('Frequency')
    plt.show()
    
    # Print descriptive statistics
    
    print_descriptive_statistics(all_details)
```

```python
analyze_data_from_json('../data/drift/milestone1-30M.json','rating', 16500, 5)
```

```python
analyze_data_from_json('../data/drift/milestone2-3M.json','rating', 16500, 5)
```

## Data Drift - Observations for Watch Time

From the observed data, there appears to be a significant shift in user engagement with the content provided by the streaming service, as evidenced by the change in watch time distribution between the old and new datasets. The transition from a unimodal to a multimodal distribution with an increased range and higher variance suggests a diversification in user preferences and content consumption patterns.

#### Changes in Central Tendency: 
The mean and median values for movie watch time in the new data (collected for milestone 2) are lower than those in the old data (milestone 1 data). This suggests that users are watching movies for shorter durations on average.


#### Changes in Distribution Shape:
The milestone1 data shows a more normal-like distribution of watch times, with a single prominent peak around the mode. The new data collected in milestone2 displays a multimodal distribution with multiple peaks.

#### Changes in Variability:
The variance is higher in the new data compared to the old data. This implies that the watch times are more spread out and there's a wider variety of movie durations being watched.

#### Possible Reasons for the drift

- There may have been changes in the user base, such as an influx of new users with different watching habits.
- Existing users might have changed their watching habits over time, possibly due to changes in their daily routines or preferences.
- External factors like holidays, global events, or changes in work-from-home policies could influence how much time users are willing to spend watching movies.
