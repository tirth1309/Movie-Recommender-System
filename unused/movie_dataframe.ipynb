```python
import pandas as pd
import json
```

```python
with open('data/small/movies_s1.json', 'r') as file:
    movie_data = json.load(file)

len(movie_data)
```

```python
# for now, can be modified later
# movie_id + user_id + original_language + popularity + runtime + vote_average

movie_dict = {
    "movie_id": [],
    "user_id": [],
    "original_language": [],
    "popularity": [],
    "runtime": [],
    "vote_average": []
}

for key, value in movie_data.items():
    for user in value['users']:
        movie_dict['movie_id'].append(key)
        movie_dict['user_id'].append(user)
        
        # below arrtibutes can be missing in json file
        try:
            movie_dict['original_language'].append(value['details']['original_language'])
        except KeyError:
            movie_dict['original_language'].append("N/A")
            
        try:
            movie_dict['popularity'].append(value['details']['popularity'])
        except KeyError:
            movie_dict['popularity'].append("N/A")
            
        try:
            movie_dict['runtime'].append(value['details']['runtime'])
        except KeyError:
            movie_dict['runtime'].append("N/A")
            
        try:
            movie_dict['vote_average'].append(value['details']['vote_average'])
        except KeyError:
            movie_dict['vote_average'].append("N/A")
```

```python
print(len(movie_dict['movie_id']))
print(len(movie_dict['user_id']))
print(len(movie_dict['original_language']))
print(len(movie_dict['popularity']))
print(len(movie_dict['runtime']))
print(len(movie_dict['vote_average']))
```

```python
movie_df = pd.DataFrame(movie_dict)
```
